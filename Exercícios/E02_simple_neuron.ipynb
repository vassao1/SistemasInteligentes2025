```
import numpy as np
from matplotlib import pyplot as plt
```

```
def sigmoid(x):
  return 1.0 / (1 + np.exp(-x))

def tanh(x):
  return np.tanh(x)
```

```
x = np.linspace(-5.0, 5.0)
y = sigmoid(x)
```

```
plt.plot(x,y)
```

```
class Neuron:
  def __init__(self, input_size=5):
    self.w = [-1,0,1]
    self.b = -0.5
  def compute(self, inputs):
    s = np.dot(self.w,inputs) + self.b
    z = tanh(s)
    return z
```

```
n = Neuron()
n1 = Neuron()
n.compute([1,2,3])
n1.compute([1,20,3])
```
